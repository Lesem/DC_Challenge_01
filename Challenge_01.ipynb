```python
import pandas as pd

tienda_1 = pd.read_csv("DC_Challenge_01/tienda_1.csv")
tienda_2 = pd.read_csv("DC_Challenge_01/tienda_2.csv")
tienda_3 = pd.read_csv("DC_Challenge_01/tienda_3.csv")
tienda_4 = pd.read_csv("DC_Challenge_01/tienda_4.csv")

ingreso_tienda_1 = tienda_1["Precio"].sum()
ingreso_tienda_2 = tienda_2["Precio"].sum()
ingreso_tienda_3 = tienda_3["Precio"].sum()
ingreso_tienda_4 = tienda_4["Precio"].sum()

ingresos = {
    "Tienda 1": ingreso_tienda_1,
    "Tienda 2": ingreso_tienda_2,
    "Tienda 3": ingreso_tienda_3,
    "Tienda 4": ingreso_tienda_4
}

ingresos_df = pd.DataFrame(list(ingresos.items()), columns=["Tienda", "Ingreso Total"])

ingresos_df.sort_values(by="Ingreso Total", ascending=False)
```

```python
tienda_1["Tienda"] = "Tienda 1"
tienda_2["Tienda"] = "Tienda 2"
tienda_3["Tienda"] = "Tienda 3"
tienda_4["Tienda"] = "Tienda 4"

df_total = pd.concat([tienda_1, tienda_2, tienda_3, tienda_4])
categorias = df_total.groupby(["Tienda", "Categoría del Producto"]).size()
categorias
```

```python
tienda_1["Tienda"] = "Tienda 1"
tienda_2["Tienda"] = "Tienda 2"
tienda_3["Tienda"] = "Tienda 3"
tienda_4["Tienda"] = "Tienda 4"

df_total = pd.concat([tienda_1, tienda_2, tienda_3, tienda_4])

resumen = df_total.groupby("Tienda").agg({
    "Precio": "sum",
    "Calificación": "mean",
    "Costo de envío": "mean"
}).reset_index()

resumen.columns = ["Tienda", "Ingresos Totales", "Calificación Promedio", "Costo Envío Promedio"]
resumen
```

```python
# 1️⃣ Gráfico de Barras – Ingresos Totales
plt.figure()
plt.bar(resumen["Tienda"], resumen["Ingresos Totales"])
plt.title("Ingresos Totales por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingresos")
plt.xticks(rotation=45)
plt.show()

#2️⃣ Gráfico de Dispersión – Ingresos vs Calificación
plt.figure()
plt.scatter(resumen["Ingresos Totales"], resumen["Calificación Promedio"])

for i in range(len(resumen)):
    plt.text(resumen["Ingresos Totales"][i],
             resumen["Calificación Promedio"][i],
             resumen["Tienda"][i])

plt.title("Relación entre Ingresos y Calificación")
plt.xlabel("Ingresos Totales")
plt.ylabel("Calificación Promedio")
plt.show()

#3️⃣ Gráfico de Líneas – Costo de Envío Promedio
plt.figure()
plt.plot(resumen["Tienda"], resumen["Costo Envío Promedio"], marker='o')
plt.title("Costo de Envío Promedio por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo Promedio de Envío")
plt.xticks(rotation=45)
plt.show()

#4️⃣ (Opcional Muy Fuerte) Categorías más vendidas
categoria_t1 = tienda_1["Categoría del Producto"].value_counts()

plt.figure()
plt.pie(categoria_t1.head(5), labels=categoria_t1.head(5).index, autopct='%1.1f%%')
plt.title("Top 5 Categorías - Tienda 1")
plt.show()
```

# **Recomendación Final**

Bajo la idea de optimizar la operación y vender la tienda menos rentable, mi recomendación es vender la "TIENDA 1"


---


**Justificación de la decisión:**

A pesar de ser la tienda con mayores ingresos brutos, es la que presenta los indicadores de salud operativa más preocupantes:

1. Es la tienda con el costo de envío promedio más alto (superior a 26,000). Esto significa que gran parte de los ingresos se están "evaporando" en gastos operativos, reduciendo el margen real de ganancia.

2. Posee la calificación promedio más baja (~3.97). Mantener esta tienda representa un riesgo para la reputación de la marca y sugiere una gestión de atención al cliente o de entrega saturada y deficiente.

3. Sus ventas dependen casi exclusivamente de dos categorías (Muebles y Electrónicos). Cualquier fluctuación en esos mercados afectaría drásticamente a esta tienda.


---



**Conclusión**

Al vender la Tienda 1, se libera de la operación más costosa y con clientes más insatisfechos.

Esto le permitirá reinvertir ese capital en las tiendas 2 o 3, que tienen costos logísticos más controlados y una mejor percepción en el mercado, logrando un negocio más ligero, rentable y fácil de administrar.
